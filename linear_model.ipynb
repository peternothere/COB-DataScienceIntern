```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
train_data = pd.read_csv("/train_dataset.csv")
train_data
```

```python
#preprocessing the training data
unnamed_columns = [col for col in train_data.columns if col.startswith("Unnamed:")]

train_data = train_data.drop(columns=unnamed_columns)
train_data
```

```python
x = train_data[['x']]
y = train_data['y']
```

```python
#trained model
model = LinearRegression()
model.fit(x, y)
```

```python
test_data = pd.read_csv("/test_dataset.csv")
test_data
```

```python
#preprocessing the test data

unnamed_columns = [col for col in test_data.columns if col.startswith("Unnamed:")]

test_data = test_data.drop(columns=unnamed_columns)

test_data
```

```python
test_data = test_data.dropna()
test_data
```

```python
x_test = test_data[['x']]
y_test = test_data['y']
```

```python
#predicted output
y_pred = model.predict(x_test)
y_pred
```

```python
plt.scatter(x_test, y_test, color='blue', label='Output')
plt.plot(x_test, y_pred, color='red', linewidth=2, label='Predicted')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

```python
r2 = r2_score(y_test, y_pred)

print("R-squared (R2) Score:", r2)
```
